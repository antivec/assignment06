# import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# generate datas

```python
num     = 201
std     = 20
a       = 2 
b       = 10
```

```python
n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b 
y2      = a * x + b

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 
# y = f(x) = a * x + b
```

# Plot the noisy data (x,y1)

```python
plt.plot(x, y1, 'b.')
plt.show()
```

# Plot the clean data (x,y2)

```python
plt.plot(x, y2, 'k--')
plt.show()
```

# means of x and y
\begin{equation*}
np.mean(x) = \overline {x} = \frac{\sum_{i=1}^n x_i}{n}
\\
np.mean(y) = \overline {y} = \frac{\sum_{i=1}^n y_i}{n}
\end{equation*}

# Find slope m 
# for the line of best fit for datas
\begin{equation*}
\ m = \frac{\sum_{i=1}^n (x_i - \overline{X})(y_i - \overline{Y})} {\sum_{i=1}^n (x_i - \overline{X})^2}
\end{equation*}

\begin{equation*}
\ k = \sum_{i=1}^n (x_i - \overline{X})(y_i - \overline{Y})
\end{equation*}

```python
k = 0
for i in range(num):
    k += (x[i] - np.mean(x)) * (y1[i] - np.mean(y1))

print(k)
```

\begin{equation*}
\ p= \sum_{i=1}^n (x_i - \overline{X})^2
\end{equation*}

```python
p = 0
for i in range(num):
    p += (x[i] - np.mean(x))*(x[i] - np.mean(x))

print(p)
```

```python
slope = k / p
print(slope)
```

# Find y-intercept of the line of best fit
\begin{equation*}
\ y-intercept = \overline{Y} - m\overline{X}
\end{equation*}

```python
y_intercept = np.mean(y1) - (slope * np.mean(x))
print(y_intercept)
```

# plot the line that fits the noisy data

```python
line = (x * slope) + y_intercept
```

```python
plt.plot(x,line,'r.',x,y1,'b.' )
plt.show()
```
